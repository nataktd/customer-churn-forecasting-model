### Алгоритм машинного обучения - Дерево Решений. Decision Tree


<font color='4682B4'>
    
Приступим к обучению модели с алгоритмом машинного обучения - `Дерево Решений`. Будем менять гиперпараметр `max_depth` и `criterion`, подберем лучшую модель.

Для оценки качества модели используем метрику `F1- мера`.
    
Создадим "универсальный" метод `learn_model_*название модели*` для каждой модели алгортмов машинного обучения. Этот метод будет принимать в качестве аргументов целевые и нецелевые признаки выборок,некоторые гиперпараметры и причие данные для заполнения накопительных таблиц.

При вызове метода модель будет подбирать гиперпараметры для самого высокого размера метрики `F1- мера`. 

Выводить самые удачные гиперпараметры, и показатели метрик:`F1- мера`, `accuracy`, `precision`, `recall`, `AUC-ROC`. Так же в качестве примечания и для сравнения выведет данные показателя класса  `DummyClassifier`, что бы показать насколько может быть обманчива точность.

Метод позволит  вывести таблицу перебора гиперпараметров для определения лучшей модели, накопительную таблицу по моделям с метриками.

Таблицу  перебора гиперпараметров для определения лучшей модели в каждом случае закоментируем, что б не заграмождать проект.
</font></div>

```python
# Cредняя точность на валидационных данных.

dummy = DummyClassifier(strategy='most_frequent')

# Метод для расчета метрик на лучшей модели    
def count_metrics(features_train, target_train, features_valid, target_valid, model_best):    
    predictions_best = model_best.predict(features_valid)
    accuracy_model_best = accuracy_score(target_valid, predictions_best).round(4)
    precision_model_best = precision_score(target_valid, predictions_best).round(4)
    recall_model_best = recall_score(target_valid, predictions_best).round(4)    
            
    # Расчет AUC-ROC, FPR, TPR
    probabilities_valid = model_best.predict_proba(features_valid)
    probabilities_one_valid = probabilities_valid[:, 1]    
    auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
    fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid)    
        
    # Cредняя точность на валидационных данных
    dummy.fit(features_train, target_train)
    dummy.predict(features_valid)
    dummy_score = dummy.score(features_valid, target_valid).round(4)
    print('Примечание: средняя точность на валидационных данных dummy', dummy_score)
    
    return predictions_best, accuracy_model_best, precision_model_best, recall_model_best, fpr, tpr, auc_roc
```

```python
# Метод для подбора гиперпараметров алгоритма Дерево Решений, вывод метрик
def learn_model_decision_tree(features_train, target_train, features_valid, target_valid, weight, class_balance, data_scaling):
    best_model_dt = None
    best_f1_dt = 0
    best_depth_dt = 0
    best_criterion_dt = ''
    depth_dt = []
    f1_dt = []
    criterion_dt = []
    
    # Перебор гиперпараметров алгоритма
    for criterion in ['gini', 'entropy']:    
        for depth in range(1,25):    
            model_dt = DecisionTreeClassifier(random_state = 123, max_depth = depth, 
                                              criterion = criterion, class_weight = weight)
            model_dt.fit(features_train, target_train)
            predictions_df = model_dt.predict(features_valid)    
            f1 = f1_score(target_valid, predictions_df).round(4)    
            if f1 > best_f1_dt:
                best_model_dt = model_dt
                best_f1_dt = f1
                best_depth_dt = depth
                best_criterion_dt = criterion        
            depth_dt.append(depth)
            f1_dt.append(f1.round(4))
            criterion_dt.append(criterion)            
        # Таблица перебора параметров
        table_dt = pd.DataFrame({'criterion': criterion_dt, 'depth': depth_dt, 'f1_score': f1_dt}) 
        
    print('Гиперпараметры и метрики по лучшей модели алгоритма машинного обучения Дерево Решений:')
    print('_________________________________')
    print('Лучший критерий для ветвления дерева -', best_criterion_dt) 
    print('Лучшая глубина дерева -', best_depth_dt) 
    print('Лучшая метрика F1- мера -', best_f1_dt.round(4))
    print('_________________________________')
    

    # Расчет метрик на лучшей модели
    model_dt_best = DecisionTreeClassifier(random_state=123, max_depth=best_depth_dt, 
                                           criterion = best_criterion_dt, class_weight = weight)
    model_dt_best.fit(features_train, target_train)    
    # вызываем метод для расчета метрик на лучшей модели
    predictions_best, accuracy_model_best, precision_model_best, recall_model_best, fpr, tpr, auc_roc = \
    count_metrics(features_train, target_train, features_valid, target_valid, model_dt_best)    
        
    # строка для накопительной таблицы по моделям и параметрам
    row_tab_dt_metrics = ['Decision_Tree', class_balance, data_scaling, best_f1_dt, 
                          accuracy_model_best, precision_model_best, recall_model_best, auc_roc]    
    
    # метод возвращает матрицу сопряженности, таблицу перебора гиперпараметров, строку с метриками для накопительной таблицы, 
    # для графика ROC-кривой параметры, лучшу модель
    return pd.crosstab(predictions_best, target_valid), table_dt, row_tab_dt_metrics, fpr, tpr, auc_roc, model_dt_best


# создаем накопительную таблицу для сбора данных по всем лучшим моделям алгоритма Деревр Решений для сравнения
tab_dt_metrics = pd.DataFrame(columns = ['Алгоритм модели', 'Классы сбалансированы', 'Данные масштабированы', 
                                         'F1- мера', 'accuracy', 'precision', 'recall', 'AUC-ROC'])
```

### Алгоритм машинного обучения - Случайный Лес. Random Forest


<font color='4682B4'>
    
Приступим к обучению модели с алгоритмом машинного обучения - `Случайный Лес`. Будем менять гиперпараметр `max_depth`, `n_estimators` и `criterion`, подберем лучшую модель.

Для оценки качества модели используем метрику `F1- мера`.
    
Создадим "универсальный" метод `learn_model_random_forest` для обучения моделей алгорима `Случайный Лес`.
</font></div>

```python
# Метод для подбора гиперпараметров алгоритма Случайный Лес, вывод метрик
def learn_model_random_forest(features_train, target_train, features_valid, target_valid, weight, class_balance, data_scaling):
    best_model_rf = None
    best_f1_rf = 0
    best_estimator_rf = 0
    best_depth_dt = 0
    depth_dt = []
    best_criterion_dt = ''
    estimator_rf = []
    f1_rf = []
    criterion_rf = []
    
    # Перебор гиперпараметров алгоритма
    for criterion in ['gini', 'entropy']:
        for depth in range(10,25):
            for estimator in range(2,15):    
                model_rf = RandomForestClassifier(random_state = 123, n_estimators = estimator, max_depth = depth, 
                                                  criterion=criterion, class_weight=weight)
                model_rf.fit(features_train, target_train)
                predictions_rf = model_rf.predict(features_valid)    
                f1 = f1_score(target_valid, predictions_rf).round(4)
    
                if f1 > best_f1_rf:
                    best_model_rf = model_rf
                    best_f1_rf = f1
                    best_estimator_rf = estimator
                    best_criterion_dt = criterion
                    best_depth_dt = depth                
        
                depth_dt.append(depth)
                estimator_rf.append(estimator)
                f1_rf.append(f1.round(4))
                criterion_rf.append(criterion)            
            # Таблица перебора параметров    
            table_rf = pd.DataFrame({'criterion': criterion_rf, 'depth': depth_dt,'estimator': estimator_rf, 'f1': f1_rf})
            
    print('Гиперпараметры и метрики по лучшей модели алгоритма машинного обучения Случайный Лес:')
    print('_________________________________')
    print('Лучший критерий для ветвления дерева -', best_criterion_dt)
    print('Количество деревьев для лучшей модели -', best_estimator_rf)
    print('Лучшая глубина дерева -', best_depth_dt)
    print('Лучшая метрика F1- мера -', best_f1_rf.round(4))
    print('_________________________________')

    # Расчет метрик на лучшей модели
    model_rf_best = RandomForestClassifier(random_state = 123, n_estimators = best_estimator_rf, max_depth = depth,
                                           criterion=best_criterion_dt, class_weight=weight)
    model_rf_best.fit(features_train, target_train)
    
    # вызываем метод для расчета метрик на лучшей модели
    predictions_best, accuracy_model_best, precision_model_best, recall_model_best, fpr, tpr, auc_roc = \
    count_metrics(features_train, target_train, features_valid, target_valid, model_rf_best)
        
    # строка для накопительной таблицы по моделям и параметрам
    row_tab_rf_metrics = ['Random_Forest', class_balance, data_scaling, best_f1_rf, 
                          accuracy_model_best, precision_model_best, recall_model_best, auc_roc]
    
    # метод возвращает матрицу сопряженности, таблицу перебора гиперпараметров, строку с метриками для накопительной таблицы, 
    # для графика ROC-кривой параметры, лучшу модель
    return pd.crosstab(predictions_best, target_valid), table_rf, row_tab_rf_metrics, fpr, tpr, auc_roc, model_rf_best

# создаем накопительную таблицу для сбора данных по всем лучшим моделям алгоритма Случайный Лес для сравнения
tab_rf_metrics = pd.DataFrame(columns = ['Алгоритм модели', 'Классы сбалансированы', 'Данные масштабированы', 
                                         'F1- мера', 'accuracy', 'precision', 'recall', 'AUC-ROC'])
```

### Алгоритм машинного обучения - Логистическая регрессия. Logistic Regression

Создадим "универсальный" метод `learn_model_logistic_regression` для обучения моделей алгорима `Логистическая Регрессия`.
</font></div>

```python
import warnings
warnings.filterwarnings('ignore')

# Метод для подбора гиперпараметров алгоритма Случайный Лес, вывод метрик
def learn_model_logistic_regression(features_train, target_train, features_valid, target_valid, weight, 
                                    class_balance, data_scaling):
    best_max_iter_lr = 0
    
    best_f1_lr = 0
    
    best_model_lr = None
    
    max_iter_lr = []
    
    f1_lr = []

    # Перебор гиперпараметров алгоритма
    for max_iter in range(10,25):
        
        model_lr = LogisticRegression(solver='lbfgs', random_state=123, 
                                      max_iter=max_iter, class_weight=weight)
        
        model_lr.fit(features_train, target_train)
        
        predictions_lr = model_lr.predict(features_valid)
        
        f1 = f1_score(target_valid, predictions_lr).round(4)
    
        if f1 > best_f1_lr:
            
            best_max_iter_lr = max_iter
            
            best_f1_lr = f1
            
            best_model_lr = model_lr

        max_iter_lr.append(max_iter)
        
        f1_lr.append(f1.round(4))
    
    # Таблица перебора параметров
    table_lr = pd.DataFrame({'max_iter': max_iter_lr, 'f1': f1_lr})
    
    print('Гиперпараметры и метрики по лучшей модели алгоритма машинного обучения Логистическая Регрессия:')
    
    print('_________________________________')
    
    print('Количество итерраций -', best_max_iter_lr)
    
    print('Лучшая метрика F1- мера -', best_f1_lr.round(4)) 

    # Расчет метрик на лучшей модели
    model_lr_best = LogisticRegression(solver='liblinear', random_state=123, 
                                       max_iter=best_max_iter_lr, class_weight=weight)
    
    model_lr_best.fit(features_train, target_train)
    
     # вызываем метод для расчета метрик на лучшей модели
    predictions_best, accuracy_model_best, precision_model_best, recall_model_best, fpr, tpr, auc_roc = \
    count_metrics(features_train, target_train, features_valid, target_valid, model_lr_best)
        
    # строка для накопительной таблицы по моделям и параметрам
    row_tab_lr_metrics = ['Logistic_Regression', class_balance, data_scaling, best_f1_lr, 
                          accuracy_model_best, precision_model_best, recall_model_best, auc_roc]
    
    # метод возвращает матрицу сопряженности, таблицу перебора гиперпараметров, строку с метриками для накопительной таблицы, 
    # для графика ROC-кривой параметры, лучшу модель
    return pd.crosstab(predictions_best, target_valid), table_lr, row_tab_lr_metrics, fpr, tpr, auc_roc, model_lr_best
```

### ROC-кривая для каждой модели. Показатель AUC-ROC


<font color='4682B4'>
    
Построим график `ROC-кривой` для каждой модели. Рассчитаем показатель `AUC-ROC`.
</font></div>

```python
# Создадим универсальный метод для визуальзации ROC-кривой
def build_roc_curve(fpr_dt, tpr_dt, fpr_rf, tpr_rf, fpr_lr, tpr_lr):
    # построим график кривой случайной модели
    plt.figure(figsize=[8,8])
    plt.plot([0, 1], [0, 1], linestyle='--', label='RandomModel')

    # ROC-кривая для модели Дерево Решений
    plt.plot(fpr_dt, tpr_dt, label='Decision Tree')
    plt.plot(fpr_rf, tpr_rf, label='Random Forest')
    plt.plot(fpr_lr, tpr_lr, label='LogisticRegression')

    plt.xlim([0,1])
    plt.ylim([0,1])

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")

    plt.legend(loc='lower right', fontsize='x-large')

    plt.title("ROC-кривая")
    plt.show()
    
    #Выведем накопительную таблицу с метриками по трем аглоритмам, сравним метрики и данные графика ROC-кривая.
    return pd.concat([tab_dt_metrics, tab_rf_metrics, tab_lr_metrics], axis=0)
```

### Метод downsampling


<font color='4682B4'>
    
Применим технику `downsampling` для нормализации баланса классов ~~и уменьшим выборку, таким образом сравняем классы целевой переменной.~~ 
</font></div>

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=123)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=123)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=123)
    
    return features_downsampled, target_downsampled

# вызываем метод на не масштабированные данные
X_train_downsampled, y_train_downsampled = downsample(X_train, y_train, 0.26)

# вызываем метод на стандартизированные данные
X_train_scaled_downsampled, y_train_downsampled = downsample(X_train_scaled, y_train, 0.26)

# вызываем метод на  нормализованные данные
X_train_scaled_mms_downsampled, y_train_downsampled = downsample(X_train_scaled_mms, y_train, 0.26)

print(y_train_downsampled.value_counts())
```

### Метод upsampling

<font color='4682B4'>
    
Применим технику `upsampling`, создадим новый `train` датасет методом `upsampling`:

- возьмём все данные с классом 1

- продублируем его N раз

- присоединим к данным класса 0 продублированный класс 1 и перемещаем, таким образом количественно сравняем классы целевой переменной. </font></div>

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

# вызываем метод на не масштабированные данные
X_train_upsampled, y_train_upsampled = upsample(X_train, y_train, 4)

# вызываем метод на  стандартизированные данные
X_train_scaled_upsampled, y_train_upsampled = upsample(X_train_scaled, y_train, 4)

# вызываем метод на нормализированные данные
X_train_scaled_mms_upsampled, y_train_upsampled = upsample(X_train_scaled_mms, y_train, 4)

print(y_train_upsampled.value_counts())
```

## Проверка моделей на тестовой выборке.

Проверим на тестовой выборке по две лучшие модели машинного обучения алгоритмов `Случайный Лес` и `Дерево Решений`:

- модель, обученная на алгоритме машинного обучение `Дерево Решений` при взвешивании классов на не масштабированных данных (номер в таблице 3 Decision_Tree) и на увеличенной выборке на масштабированных данных (номер в таблице 9 Decision_Tree),

- модель, обученная на алгоритме машинного обучение `Случайный Лес` при увеличении выборки на не масштабированных данных (номер в таблице 8 Random_Forest) и на масштабированных данных (номер в таблице 9 Random_Forest). </font></div>

```python
# Метод для проверки тестовой выборки и  вывода метрик
def predict_test(features_test, target_test, best_model, algorithm, class_balance, data_scaling):
    predictions_test = best_model.predict(features_test)
    accuracy_model_best = accuracy_score(target_test, predictions_test).round(4)
    precision_model_best = precision_score(target_test, predictions_test).round(4)
    recall_model_best = recall_score(target_test, predictions_test).round(4)    
    probabilities_test = best_model.predict_proba(features_test)
    probabilities_one_test = probabilities_test[:, 1]    
    auc_roc = roc_auc_score(target_test, probabilities_one_test).round(4)
    fpr, tpr, thresholds = roc_curve(target_test, probabilities_one_test)    
    best_f1 = f1_score(target_test, predictions_test).round(4)
    
        
    # строка для накопительной таблицы по моделям и параметрам
    row_tab_metrics = [algorithm, class_balance, data_scaling, best_f1, 
                          accuracy_model_best, precision_model_best, recall_model_best, auc_roc]
        
    # метод возвращает матрицу сопряженности, строку с метриками для накопительной таблицы, 
    # для графика ROC-кривой параметры
    return pd.crosstab(predictions_test, target_test), row_tab_metrics, fpr, tpr, auc_roc


# создаем накопительную таблицу для сбора данных по метрикам для сравнения
total_tab_metrics = pd.DataFrame(columns = ['Алгоритм модели', 'Классы сбалансированы', 'Данные масштабированы', 
                                         'F1- мера', 'accuracy', 'precision', 'recall', 'AUC-ROC'])
```

```python
# создадим словарь для перебора параметров моделей
params_model = {model_dt_3: ['Decision_Tree', 'взвешенные классы', 'нет', X_test],
                model_dt_9: ['Decision_Tree', 'увеличение выборки', 'стандарт', X_test_scaled],
                model_rf_8: ['Random_Forest', 'увеличение выборки', 'нет', X_test],
                model_rf_9: ['Random_Forest', 'увеличение выборки', 'стандарт', X_test_scaled]}

plt.figure(figsize=[10,8])
# построим график кривой случайной модели
plt.plot([0, 1], [0, 1], linestyle='--', label='RandomModel')    

for key, value in params_model.items():
    best_model = key
    algorithm = value[0]
    class_balance = value[1]
    data_scaling = value[2]
    X_test = value[3]
    cross_tab, total_tab_metrics.loc[len(total_tab_metrics.index)], fpr, tpr, auc_roc =\
    predict_test(X_test, y_test, best_model, algorithm, class_balance, data_scaling )
        
    # ROC-кривая для модели 
    plt.plot(fpr, tpr, label=algorithm + ' ' + class_balance)

plt.xlim([0,1])
plt.ylim([0,1])

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.legend(loc='lower right', fontsize='x-large')

plt.title("ROC-кривая")
plt.show()

# вызовем накопительную таблицу для сбора данных по метрикам для сравнения    
total_tab_metrics
```
